### 3.3 complications of concurrent processing

1. access to data has to be synchronized so 2 processes don't write to same address space simultaneously
2. processes have to be scheduled and a timer set so that CPU cycles are allotted optimally and no process takes up too much time.
3.  OS has to ensure that each process only accesses data it is supposed to
4.  Continuous communication between OS and PCB to locate and maintain register values, ie. the context is located in the PCB and loaded to registers, sometimes multiple if available -> can overload system.
5.  Processes that require large data blocks for execution can enter deadlocks waiting for resources to be freed (isn't this what the scheduler handles?)  To balance out heavy IO processes and heavy CPU processes in the wait queue or other queue.



## IPC in Posix with shared memory area (Producer Consumer)
see C files share_mem.c 

```c
#include <stdio.h>
#include <stdlib.h>
#include <string.h>
#include <fcntl.h>
#include <sys/shm.h>
#include <sys/stat.h>
#include <sys/mman.h>

int main(){
    /* the size (in bytes) of shared memory object */
    const int SIZE = 4096;
    /* name of the shared memory object */
    const char *name = "OS";
    /* strings written to shared memory */
    const char *message_0 = "Hello";
    const char *message_1 = "World!";
    /* shared memory file descriptor */
    int fd;
    /* pointer to shared memory obect */
    char *ptr; 
    char *original;
    /* create the shared memory object */
    fd = shm_open(name,O_CREAT | O_RDWR,0666);
    /* configure the size of the shared memory object */
    ftruncate(fd, SIZE);
    /* memory map the shared memory object */
    ptr = (char *) mmap(0, SIZE, PROT_READ | PROT_WRITE, MAP_SHARED, fd, 0);
    original = ptr;
    /* write to the shared memory object */
    sprintf(ptr,"%s",message_0);
    ptr += strlen(message_0);
    sprintf(ptr,"%s",message_1);
    ptr += strlen(message_1);
    
    return 0;
}
```

### Pipes for message passing usually from parent to child, parent creates, child inherits.


fd[1] is the write end and fd[0] is the read end.  Pipes in Linux are unidirectional unlike the diagram below:  http://man7.org/linux/man-pages/man7/pipe.7.html  Treated as a special type of file so ordinary system calls read() and write() access pipe.  

```c
#include <sys/types.h>
#include <stdio.h>
#include <string.h>
#include <unistd.h>

#define BUFFER_SIZE 25
#define READ_END 0
#define WRITE_END 1

int main(void)
{
    char write_msg[BUFFER_SIZE] = "Greetings";
    char child_msg[BUFFER_SIZE] = "aloha!";
    char read_msg[BUFFER_SIZE];
    char parent_read[BUFFER_SIZE];
    int fd[2]; //0 and 1
    pid_t pid;
    
    /* create the pipe */
    if (pipe(fd) == -1) {
        fprintf(stderr,"Pipe failed");
        return 1;
    }
    /* fork a child process */
    pid = fork();
    
    if (pid < 0) { /* error occurred */
        fprintf(stderr, "Fork Failed");
        return 1;
    }
    
    if (pid > 0) { /* parent process */
        
        /* write to the pipe */
        write(fd[WRITE_END], write_msg, strlen(write_msg)+1);

        /* close the write end of the pipe, this signals EOF returns 0 to read */
        close(fd[WRITE_END]);
        
        //parent reads msg
        //read(fd[READ_END], parent_read, BUFFER_SIZE);
        
        /* close the unused end of the pipe */
        close(fd[READ_END]);
    
    }
    else { /* child process */
        //child writes message
       // write(fd[WRITE_END], child_msg, strlen(child_msg)+1); //+1 for close null signal
        /* close the unused end of the pipe */
        close(fd[WRITE_END]);
    
        /* read from the pipe */
        read(fd[READ_END], read_msg, BUFFER_SIZE);
        printf("read %s",read_msg);

        /* close the read end of the pipe */
        close(fd[READ_END]);
    }
    return 0;
}

```
